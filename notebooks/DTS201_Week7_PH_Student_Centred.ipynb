## 4. One-Sample t-Test

A **one-sample t-test** compares the mean of a sample to a known or hypothesized value.

Example question:

> â€œIs the **average exam score** significantly different from 50?â€

Steps:

1. Choose a **numerical variable** (e.g. `Exam_Score`).  
2. Specify a **hypothesized mean** (e.g. 50).  
3. Run a one-sample t-test and interpret the **p-value**.


```
# Choose a numerical column for the one-sample t-test
# Replace 'Exam_Score' with a real numerical column from your dataset
score_col = "Exam_Score"  # TODO: change this to a real numeric column

# Set the hypothesized population mean (e.g. pass mark, expected average)
mu_0 = 50  # TODO: change this to a meaningful hypothesized value

# Drop missing values from the selected column to avoid errors in the t-test
scores = df[score_col].dropna()  # Create a clean Series without NaN values

# Perform one-sample t-test: H0: mean = mu_0
t_statistic, p_value = stats.ttest_1samp(scores, popmean=mu_0)

# Print the test results
print(f"One-sample t-test for mean of {score_col} vs {mu_0}")
print(f"t-statistic: {t_statistic:.3f}")
print(f"p-value: {p_value:.5f}")
```

## 5. Independent Samples t-Test (Two Groups)

An **independent samples t-test** compares the means of a numerical variable between **two independent groups**.

Example question:

> â€œDo **male** and **female** students have different average exam scores?â€

Steps:

1. Choose a **categorical column** with exactly two groups (e.g. `Gender`: Male/Female).  
2. Choose a **numerical outcome variable** (e.g. `Exam_Score`).  
3. Run the test and interpret the p-value.


```
# Choose the grouping (categorical) column and the numerical outcome
# Replace 'Gender' and 'Exam_Score' with relevant columns from your dataset
group_col = "Gender"      # TODO: change to your binary categorical column
numeric_col = "Exam_Score"  # TODO: change to your numeric outcome column

# Inspect the unique groups in the grouping column
df[group_col].value_counts()  # Shows the categories and their counts
```

```
# Extract data for the two groups (e.g. Male and Female)
# Replace 'Male' and 'Female' with the actual category names in your data
group1_label = "Male"    # TODO: set first group label
group2_label = "Female"  # TODO: set second group label

# Select numerical values for each group, dropping missing values
group1_values = df[df[group_col] == group1_label][numeric_col].dropna()
group2_values = df[df[group_col] == group2_label][numeric_col].dropna()

# Display basic descriptive statistics for each group
print(f"Group {group1_label} - mean:", group1_values.mean(), ", n =", len(group1_values))
print(f"Group {group2_label} - mean:", group2_values.mean(), ", n =", len(group2_values))
```

```
# Perform an independent samples t-test (assuming equal variances by default)
t_stat, p_val = stats.ttest_ind(group1_values, group2_values, equal_var=True)

# Print the t-test results
print(f"Independent samples t-test: {numeric_col} by {group_col}")
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_val:.5f}")
```

## 6. Chi-Square Test of Independence

A **chi-square test of independence** is used to check whether there is an association between **two categorical variables**.

Example questions:

- â€œIs there an association between **Gender** and **Pass/Fail**?â€  
- â€œIs there a relationship between **Programme** and **Satisfaction Level** (Low/Medium/High)?â€  

We first build a **contingency table** (cross-tabulation) and then apply the chi-square test.


```
# Choose two categorical variables for chi-square test
# Replace 'Gender' and 'Pass_Fail' with real categorical columns
cat1 = "Gender"     # TODO: change this to first categorical variable
cat2 = "Pass_Fail"  # TODO: change this to second categorical variable

# Create a contingency table (cross-tabulation) of the two variables
contingency_table = pd.crosstab(df[cat1], df[cat2])

# Display the contingency table
contingency_table
```

```
# Perform chi-square test of independence using scipy
chi2_stat, p_chi, dof, expected = stats.chi2_contingency(contingency_table)

# Print the test results
print("Chi-square test of independence between", cat1, "and", cat2)
print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"Degrees of freedom: {dof}")
print(f"p-value: {p_chi:.5f}")

# Convert expected frequencies to a DataFrame for easier viewing
expected_freq = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
print("
Expected frequencies under independence:")
expected_freq
```

## 7. Correlation Analysis

**Correlation** measures the strength and direction of the **linear relationship** between two numerical variables.

- Correlation coefficient (r) ranges from **-1 to +1**.  
- **r > 0** â€“ positive relationship (as one increases, the other tends to increase).  
- **r < 0** â€“ negative relationship (as one increases, the other tends to decrease).  
- **r â‰ˆ 0** â€“ weak or no linear relationship.


```
# Choose two numerical variables for correlation analysis
# Replace 'Hours_Studied' and 'Exam_Score' with real numeric columns
num1 = "Hours_Studied"  # TODO: change to first numeric variable
num2 = "Exam_Score"     # TODO: change to second numeric variable

# Drop rows where either variable is missing
pair_data = df[[num1, num2]].dropna()  # Keep only rows with non-missing values for both columns

# Compute Pearson correlation coefficient and p-value
r_value, p_corr = stats.pearsonr(pair_data[num1], pair_data[num2])

# Print correlation results
print(f"Pearson correlation between {num1} and {num2}:")
print(f"r = {r_value:.3f}")
print(f"p-value = {p_corr:.5f}")
```

```
# Visualize the relationship between the two variables with a scatter plot and trend line
plt.figure(figsize=(7, 5))  # Set figure size

# Create scatter plot
sns.scatterplot(x=pair_data[num1], y=pair_data[num2])  # Each point is one observation

# Add a regression (trend) line using seaborn's regplot
sns.regplot(x=num1, y=num2, data=pair_data, scatter=False, color="red")  # Only draws the line, no extra points

# Add title and labels
plt.title(f"Scatter Plot and Trend Line: {num1} vs {num2}")
plt.xlabel(num1)
plt.ylabel(num2)

# Display the plot
plt.tight_layout()
plt.show()
```
